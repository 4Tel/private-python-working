```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('classic')
plt.rcParams['figure.dpi'] = 200
```

```python
# Example Data
x = np.linspace(0, 10, 500)
y1 = np.sin(x)
y2 = np.sin(x) + 0.05

# inset range
x1, x2 = 1.2, 1.8
y1_min = np.min([y1[(x>=x1) & (x<=x2)], y2[(x>=x1) & (x<=x2)]])
y1_max = np.max([y1[(x>=x1) & (x<=x2)], y2[(x>=x1) & (x<=x2)]])
```

```python
# Total Range Graph
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x, y1, label='Graph 1')
ax.plot(x, y2, label='Graph 2')
ax.legend()
ax.set_title('Total Range Graph')

# Inset (Zoomed) Graph
pos_x = 0.4
pos_y = 0.5
multiply_x = 4
multiply_y = 4
xlim = ax.get_xlim()
ylim = ax.get_ylim()
x_port = (x2 - x1)/(xlim[1] - xlim[0]) * multiply_x
y_port = (y1_max - y1_min)/(ylim[1] - ylim[0]) * multiply_y
axins = ax.inset_axes([pos_x, pos_y, x_port, y_port])  # [x, y, width, height] in figure coordinates
axins.plot(x, y1)
axins.plot(x, y2)
axins.set_xlim(x1, x2)
axins.set_ylim(y1_min, y1_max)
axins.set_xticks([])
axins.set_yticks([])

# Zoomed area display
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
mark_inset(ax, axins, loc1=1, loc2=3, fc="none", ec="0.5")

plt.show()
```
