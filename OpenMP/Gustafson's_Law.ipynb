```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../')
import plotForm
plotForm.configure_matplotlib()

plt.title("Gustafson's Law")
plt.xlabel("Number of Processors (N)")
plt.ylabel("Speedup (S)")

P_vals = np.linspace(0.1,0.9,9)
N = np.arange(0,101)
P_vals = P_vals[::-1]
for P in P_vals:
    speedup = (1-P) + N*P
    plt.plot(N, speedup, lw=5, label=f'{P:.0%}')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.legend(title='Parallel Portion (P)', loc='center left', 
           **plotForm.legend_params, bbox_to_anchor=(1, 0.5))
plt.show()
```
