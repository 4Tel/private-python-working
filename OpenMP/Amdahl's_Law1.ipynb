```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../')
import plotForm
plotForm.configure_matplotlib()

plt.title('Amdahl\'s Law')
plt.xlabel('Number of Processors (N)')
plt.ylabel('Speedup (S)')

def amdahls_law(P, N):
    return N and 1 / ((1 - P) + P/N)

P_values = [0.25, 0.5, 0.9, 0.95]
N_values = np.arange(0, 101)

for P in P_values:
    speedups = [amdahls_law(P, N) for N in N_values]
    plt.plot(N_values, speedups, label=f'{P:.0%}')

plt.xlim(0, 100)
plt.legend(title="Parallel Portion (P)",loc='best',**plotForm.legend_params)
plt.tight_layout()
plt.show()
```
