```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../')
import plotForm
plotForm.configure_matplotlib()

plt.title("Amdahl's Law")
plt.xlabel("Parallel Fraction")
plt.ylabel("Speedup")

def amdahls_law(P, N):
    return 1 / ((1 - P) + (P / N))

P = np.linspace(0, 1, 100)  # 병렬 비율 (0부터 1까지)
N = 10  # 사용할 코어 수

speedups = [amdahls_law(f, N) for f in P]

# 그래프 그리기
plt.plot(P * 100, speedups, label=f'{N} Cores', color='blue')
plt.xlabel("Parallel Fraction (%)")
plt.xlim(0, 100)
plt.legend(loc='best',fontsize=18,**plotForm.legend_params)
plt.tight_layout()
plt.show()
```
