```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6), dpi=300)
plt.style.use('default')
plt.rcParams.update({'font.size': 14})
plt.tick_params(axis='both', direction='in', top=True, right=True)
plt.title("Compensated Amdahl's Law")
plt.xlabel("Number of Processors")
plt.ylabel("Speedup")

# Parameters
N = np.arange(1, 80)  # Number of processors

# Plot multiple scenarios
scenarios = [
    {'P': 0.9, 'k': 0.001, 'color': 'blue'},
    {'P': 0.99, 'k': 0.01, 'color': 'red'}
]

for scenario in scenarios:
    P = scenario['P']
    k = scenario['k']
    # Compensated Amdahl's Law formula
    speedup = 1 / ((1 - P) + (P/N) + k * (N - 1))
    plt.plot(N, speedup, label=f'P = {P:.0%}, $T_o$ = {k:.1%}', 
             color=scenario['color'], linewidth=2)

# Improved legend visibility
plt.legend(loc='best', frameon=True, fancybox=True, shadow=True, 
           fontsize=18, framealpha=0.9, edgecolor='black')
plt.grid(alpha=0.3)
plt.show()
```
