```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../')
import plotForm
plotForm.configure_matplotlib()

plt.title("Compensated Amdahl's Law")
plt.xlabel("Number of Processors (N)")
plt.ylabel("Speedup (S)")

# Parameters
N = np.arange(0, 81)  # Number of processors

# Plot multiple scenarios
scenarios = [
    {'P': 0.9, 'k': 0.001, 'color': 'blue'},
    {'P': 0.99, 'k': 0.01, 'color': 'red'}
]

for scenario in scenarios:
    P = scenario['P']
    k = scenario['k']
    # Compensated Amdahl's Law formula
    speedup = 1 / ((1 - P) + (P/N) + k * (N - 1))
    plt.plot(N, speedup, label=f'P = {P:.0%}, $T_o$ = {k:.1%}', 
             color=scenario['color'], linewidth=2)

# Improved legend visibility
plt.xlim(0, 80)
plt.legend(loc='right', fontsize=18, **plotForm.legend_params)
plt.tight_layout()
plt.show()
```
